```python
#@title Data Cleaning

# function to fix heights
def convert_height(height_str):
    if pd.isna(height_str) or height_str in ['-', 'None', 'So', 'Jr', 'Fr', '0']:
        return None
    elif height_str == 'Apr-00':
        return 4*12
    elif height_str == 'May-00':
        return 5*12
    elif height_str == 'Jun-00':
        return 6*12
    elif height_str == 'Jul-00':
        return 7*12
    elif "'" in height_str:
        feet, inches = height_str.split("'")
        return int(feet) * 12 + int(inches)
    elif '-' in height_str:
        try:
            inch, ft = height_str.split("-")
            if ft == 'Jun':
              return int(inch) + 6*12
            elif ft == 'Jul':
              return int(inch) + 7*12
            elif ft == 'May':
              return int(inch) + 5*12
            elif ft == 'Apr':
              return int(inch) + 4*12
        except ValueError:
            return height_str
    else:
        return height_str

# Apply the function to the 'ht' column
train['ht'] = train['ht'].apply(convert_height)

# Apply the function to the 'ht' column
test['ht'] = test['ht'].apply(convert_height)
```

```python
#@title Undersampling (Filter on 'pick')

train2 = train[~train['pick'].isna()]
train2.head()
```

```python
#@title Feature Selection

df_cleaned = train2.drop(columns = ['type','player_id', 'num'], axis=1)
```

```python
#@title Filling N/As

df_cleaned['Rec_Rank'] = df_cleaned['Rec_Rank'].fillna(0)
df_cleaned['pick'] = df_cleaned['pick'].fillna(0)

test['Rec_Rank'] = test['Rec_Rank'].fillna(0)
#test['pick'] = test['pick'].fillna(0)
```

```python
#@title Modelling

from pycaret.classification import *

setup(df_cleaned, target='drafted', categorical_features=['team','conf','year','yr', 'pick'], imputation_type='simple', numeric_imputation='median') #, feature_selection = True, feature_selection_estimator='lr')

best_model = compare_models()
```

```python
lr = create_model('lr')
```

```python
evaluate_model(lr)
```

```python
tunedlr = tune_model(lr, optimize = 'AUC')
```

```python
evaluate_model(tunedlr)
```

```python
#@title Predictions

predictions = predict_model(tunedlr, data=test, raw_score=True)
predictions['drafted'] = predictions['prediction_score_1']
predictions['drafted'] = np.where(predictions['pick'].isna(), 0, predictions['drafted'])
predictions[['player_id', 'drafted']].to_csv('pred_2508.csv', index=False)
```
